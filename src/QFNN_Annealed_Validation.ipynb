# ðŸŒŒ QFNN Annealed Training Validation

Analyze and visualize the behavior of the Simulated Annealing patched QFNN model after training.

```python
# ðŸ”§ Load modules
import torch
import matplotlib.pyplot as plt
import os

# Load saved training history
history = torch.load('C:/Users/casma/OneDrive/Desktop/physicsFirst/quantum_field_engine/trained_model/training_history.pt')
```

```python
# ðŸ“Š Plot full training evolution
steps = range(len(history['loss']))
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

ax[0, 0].plot(steps, history['loss'])
ax[0, 0].set_title("Total Loss over Steps")

ax[0, 1].plot(steps, history['phase_loss'])
ax[0, 1].set_title("Phase Loss over Steps")

ax[1, 0].plot(steps, history['entropy'])
ax[1, 0].set_title("Entropy over Steps")

ax[1, 1].plot(steps, history['coherence'])
ax[1, 1].set_title("Coherence over Steps")

for a in ax.flat:
    a.set_xlabel("Steps")
    a.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
