
# 🌌 QFNN: Quantum Field Neural Network Showcase

This notebook demonstrates a novel symbolic engine based on quantum field evolution.  
It visualizes how meaning, coherence, and memory are physically encoded in the QFNN field space.

Built using golden-ratio phase alignment, simulated annealing, and Hebbian field memory.


```plaintext

import torch
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from axiomatic_qfnn import AxiomaticQuantumField

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
```

## 🔧 Field Initialization

```plaintext

model = AxiomaticQuantumField()
TEXT = "knowledge is light and understanding is illumination"
tokens = model.tokenize(TEXT)
field = model.init_field(tokens)

# Initial radial-phase plot
r = field["r"].cpu().numpy()
θ = field["θ"].cpu().numpy()
plt.figure(figsize=(6, 6))
plt.polar(θ, r, marker='o')
plt.title("Initial Token Phase Field (Radial Spiral Init)")
plt.show()
```

## 🔁 Field Evolution (Annealing + Memory)

```plaintext

θ_target = torch.roll(field["θ"], shifts=-1)
θ_history = []
coherence_history = []

for step in range(model.config["steps"]):
    field, loss, phase_loss, entropy, coherence = model.heun_step(field, θ_target, step)
    θ_history.append(torch.angle(field["ψ"]).cpu().numpy())
    coherence_history.append(coherence.item())
    
plt.plot(coherence_history)
plt.title("Coherence over Time")
plt.xlabel("Steps")
plt.ylabel("Coherence")
plt.grid(True)
plt.show()
```

## 📊 Final Phase Field in 3D

```plaintext

ψ = field["ψ"].cpu()
r = torch.abs(ψ).numpy()
θ = torch.angle(ψ).numpy()
delta = np.abs((θ - np.roll(θ, -1) + np.pi) % (2 * np.pi) - np.pi)

x = r * np.cos(θ)
y = r * np.sin(θ)
z = delta
tokens = tokens

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(x, y, z, c=z, cmap="plasma", s=80)
for i, token in enumerate(tokens):
    ax.text(x[i], y[i], z[i], token, fontsize=9)

ax.set_title("🌌 Token Field in Phase-Radius-Error Space")
ax.set_xlabel("r·cos(θ)")
ax.set_ylabel("r·sin(θ)")
ax.set_zlabel("Δθ")
plt.colorbar(sc, label="Phase Error")
plt.tight_layout()
plt.show()
```

## 🧠 Hebbian Memory Field

```plaintext

W = field["W"].cpu().abs().numpy()
plt.figure(figsize=(8, 6))
plt.imshow(W, cmap='inferno')
plt.colorbar(label="|W_ij|")
plt.title("Hebbian Memory Field (|W|)")
plt.xlabel("Token j")
plt.ylabel("Token i")
plt.show()
```
